```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Sample data banaate hain
np.random.seed(0)
X = np.random.rand(100) * 10  # Independent variable
Y = 2 * X + 3 + np.random.randn(100)  # Dependent variable with noise

# Function to fit and plot regression
def fit_and_plot(X, Y, title):
    X_with_const = sm.add_constant(X)  # Intercept ke liye constant jodte hain
    model = sm.OLS(Y, X_with_const).fit()
    print(f"{title} Regression Coefficients:")
    print(f"Intercept: {model.params[0]}, Slope: {model.params[1]}")
    
    plt.scatter(X, Y, color='blue', label='Data Points')
    plt.plot(X, model.predict(X_with_const), color='red', label='Regression Line')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()

# Original regression
fit_and_plot(X, Y, 'Original Data')

# Change of Origin (Y mein 5 jodna)
Y_shifted = Y + 5
fit_and_plot(X, Y_shifted, 'Y Shifted by 5 (Change of Origin)')

# Change of Scale (Y ko 2 se guna karna)
Y_scaled = Y * 2
fit_and_plot(X, Y_scaled, 'Y Scaled by 2 (Change of Scale)')
```

# **`Overall`**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate the data
x = 10 * np.random.rand(50)
y = 3 * x - 8 + np.random.randn(50) * 4

# Fit a linear regression model
x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)

# Calculate the predicted values
y_pred = model.predict(x)

# Plot the scatter plot and regression lines
plt.scatter(x, y, label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot with increased variability and regression lines")

# Plot the actual population line
x_line = np.linspace(0, 10, 100)
y_actual = 2 * x_line - 5
plt.plot(x_line, y_actual, 'r', label="Population line (m=3, b=-8)")

# Plot the estimated regression line
y_estimated = model.coef_[0] * x_line + model.intercept_
plt.plot(x_line, y_estimated, 'g', label=f"Estimated line (m={model.coef_[0]:.2f}, b={model.intercept_:.2f})")

# Add legend and show the plot
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
```

**` np.random.seed() ensures that anyone running your code gets the same results as you.
. You can set it to any integer value. The reason 42 is often used in examples is due to its reference in popular culture`**

```python
# Generate synthetic data
np.random.seed(42) 
n = 100 # number of data points
```

**`x1 and x2 are generated from a normal (Gaussian) distribution with a mean of 0 and a standard deviation of 1.
Both x1 and x2 contain n = 100 random values, representing two independent predictor variables.`**

```python
x1 = np.random.normal(0,1,n)
x2 = np.random.normal(0,1,n)
```

**`This creates 10 additional irrelevant predictors that are also generated from a normal distribution with mean 0 and standard deviation 1.
These predictors are stored in a 2D array of shape (n, 10), meaning there are n = 100 samples and 10 irrelevant predictors. They are not related to the response variable y.`**

```python
irrelevant_predictors = np.random.normal(0,1,(n,10))
```

`This line generates the response variable y, which is dependent on the predictors x1 and x2.`


```python
y = 2* x1 + 3 * x2 + np.random.normal(0,1,n)
```

```python
# Helper function to calculate adjusted R-squared
def adjusted_r2(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)
```

### `Fit linear regression models with different predictors`


```python
x = pd.DataFrame({'x1': x1, 'x2': x2})
x
```

**`This part generates a list of 10 pd.Series objects from the irrelevant_predictors array.
irrelevant_predictors[:, i]: This selects the i-th column (predictor) from the irrelevant_predictors matrix.
pd.Series(irrelevant_predictors[:, i], name=f"irrelevant_{i}"): This converts the selected column into a Pandas Series and gives it a name like "irrelevant_0", "irrelevant_1", ..., "irrelevant_9", making the column names more meaningful.`**

```python
x_with_irrelevant = pd.concat([x] + [pd.Series(irrelevant_predictors[:,i], name=f"irrelevant_{i}") for i in range(10)], axis =1)
```

```python
x_with_irrelevant
```

```python

model_1 = LinearRegression().fit(x,y)
model_1
```

```python
model_2 = LinearRegression().fit(x_with_irrelevant,y)
model_2
```

# Calculate R-squared and adjusted R-squared for each model

```python
print(x.shape)
x.shape[1]
```

```python
print(x_with_irrelevant.shape)
x_with_irrelevant.shape[1]
```

```python
models = [('Model with relevant predictors', model_1,x.shape[1]),('Model with irrelevant predictors',model_2, x_with_irrelevant.shape[1])]
models
```

```python
for name, model, k in models:
    r2 = r2_score(y, model.predict(x_with_irrelevant.iloc[:, :k]))
    adj_r2 = adjusted_r2(r2, n, k)
    print(f"{name}: R-squared = {r2:.3f}, Adjusted R-squared = {adj_r2:.3f}")
```

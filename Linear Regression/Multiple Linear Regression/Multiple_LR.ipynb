```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('50_Startups.csv')
df
```

```python
x = df.iloc[:,0:df.shape[1]-1]
x
print(type(x))
```

```python
y=df.iloc[:,df.shape[1]-1]
y
```

```python
states = pd.get_dummies(x['State'],drop_first=True) #Convert categorical variable to numerical using one hot encoding
```

```python
states=states.astype(int)
```

```python
states
```

```python
x
```

```python
x=x.drop('State',axis=1) #axis=1 for column and axis=0 for row
x
```

```python
x = pd.concat([x,states],axis=1) 
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
type(x_test)
```

```python
y_test
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score # if our r2_score near to 1 it is very good
```

```python
y_pred.shape
```

# **`Visualize Actual and Predicted Values`**

```python
plt.plot(y_test,y_pred,'o') # y_test-> jo hmara actual output h , y_pred-> jo model output diya h
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual and Predicted Values')
```

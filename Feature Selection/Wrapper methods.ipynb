### What is 5-Fold Cross-Validation?
5-Fold Cross-Validation splits the dataset into 5 equal parts (or as equal as possible).
The model is trained on 4 parts (training set) and tested on the remaining 1 part (validation set).\
This process is repeated 5 times, with each fold taking a turn as the validation set, while the remaining 4 folds are used for training.\
After all 5 iterations, the average performance (accuracy, precision, etc.) is calculated over the 5 validation sets.
### Why Use Cross-Validation?
**Better Performance Estimate:** Instead of relying on a single train-test split, cross-validation ensures that every observation in the dataset gets used for both training and validation. This reduces the likelihood of overfitting or underfitting because the model is tested multiple times.\
**Robustness:** It gives a better sense of how the model will generalize to new data, especially when the dataset is small.\
**Model Selection:** In feature selection, cross-validation helps ensure that the chosen features perform well across different subsets of data, leading to more reliable and generalizable feature choices.

# Regression Example

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df['medv'], test_size=0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.head()
```

```python
# This is for sare data ko same scale pe lane k liye
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# baseline model means witout FS accuracy check krenge
import numpy as np
from sklearn.metrics import r2_score
model = LinearRegression()


print("training",np.mean(cross_val_score(model, X_train, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test, y_test, cv=5, scoring='r2')))
```

```python
lr = LinearRegression()

exh = EFS(lr, max_features=13, scoring='r2', cv=10, print_progress=True,n_jobs=-1) # n_jobs=-1 for fast execution

sel = exh.fit(X_train, y_train)
```

```python
# See Feature:8191/8191-> 2^13-1 subset generatedted
```

```python
sel.best_score_
```

```python
sel.best_feature_names_ # ye subset sbse acha perform kr rha h
```

```python
metric_df = pd.DataFrame.from_dict(sel.get_metric_dict()).T
metric_df
```

```python
# bcz no. of features jada h hme r2_score pe bhrosha ni hai
def adjust_r2(r2, num_examples, num_features):
    coef = (num_examples - 1) / (num_examples - num_features - 1) 
    return 1 - (1 - r2) * coef
```

```python
metric_df['observations'] = 8191
metric_df['num_features'] = metric_df['feature_idx'].apply(lambda x:len(x))
metric_df['adjusted_r2'] = adjust_r2(metric_df['avg_score'],metric_df['observations'],metric_df['num_features'])
```

```python
metric_df.sort_values('adjusted_r2',ascending=False)
```

```python
X_train_sel = sel.transform(X_train)
X_test_sel = sel.transform(X_test)

model = LinearRegression()

print("training",np.mean(cross_val_score(model, X_train_sel, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test_sel, y_test, cv=5, scoring='r2')))
```

# **'Sequential Backward Elimination'**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
data
# separate the target variable
X = data.drop("medv", axis=1)
y = data['medv']

# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print(X_train.shape)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = LinearRegression()

print("training",np.mean(cross_val_score(model, X_train, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test, y_test, cv=5, scoring='r2')))
```

```python
lr = LinearRegression()

# perform backward elimination
sfs = SFS(lr, k_features='best', forward=False, floating=False, scoring='r2',cv=5) # forward ko false krne pe backward selection and True krne pe forward selection ho jata h

sfs.fit(X_train, y_train)
```

```python
sfs.k_feature_idx_
```

```python
metric_df = pd.DataFrame.from_dict(sfs.get_metric_dict()).T

metric_df['observations'] = 404
metric_df['num_features'] = metric_df['feature_idx'].apply(lambda x:len(x))
metric_df['adjusted_r2'] = adjust_r2(metric_df['avg_score'],metric_df['observations'],metric_df['num_features'])

metric_df
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs


fig1 = plot_sfs(sfs.get_metric_dict(), kind='std_err',)

plt.title('Sequential Backward Elimination (w. StdErr)')
plt.grid()
plt.show()
```

```python
X_train_sel = sfs.transform(X_train)
X_test_sel = sfs.transform(X_test)

model = LinearRegression()

print("training",np.mean(cross_val_score(model, X_train_sel, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test_sel, y_test, cv=5, scoring='r2'))) 
```

# `Sequential Forward Elimination`

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')

# separate the target variable
X = data.drop("medv", axis=1)
y = data['medv']

# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print(X_train.shape)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
